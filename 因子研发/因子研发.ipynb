## 步骤四： 计算因子（数据或算法）
算法：add_formula()

数据：dv.get_ts()

### 计算聪明度因子

```python
dv.add_formula('smart_degree','Abs(Return(close_adj,25))/Log(Ewma(volume,25))',is_quarterly=False,add_data=True)
dv.get_ts('smart_degree').tail()
```

### 计算非流动性因子

```python
N=30
dv.add_formula("ILLIQ","(Ts_Mean((Return(close_adj,1)/(float_mv)),%s))"%(N),is_quarterly=False,add_data=True)
dv.get_ts('ILLIQ').tail()
```

### 计算bias_std_turn_Nm因子

```python
dv.add_formula('bias_std_turn_Nm','(StdDev(turnover,30)/StdDev(turnover,252)-1)',is_quarterly=False,add_data=True)
dv.get_ts('bias_std_turn_Nm').tail()
```

### 计算bias_turn_Nm因子

```python
dv.add_formula('bias_turn_Nm','Ts_Mean(turnover,1)/Ts_Mean(turnover,252)-1',is_quarterly=False,add_data=True)
dv.get_ts('bias_turn_Nm').tail()
```

### 计算std_turn_Nm因子

```python
dv.add_formula('std_turn_Nm','StdDev(turnover,30)',is_quarterly=False,add_data=True)
dv.get_ts('std_turn_Nm').tail()
```

### 计算high_r_std_Nm因子

```python
dv.add_formula('high_r_std_Nm','StdDev(high_adj/Delay(close_adj,1),30)',is_quarterly=False,add_data=True)
dv.get_ts('high_r_std_Nm').tail()
```

### 计算hml_r_std_Nm因子

```python
dv.add_formula('low_r_std_Nm','StdDev(low_adj/Delay(close_adj,1),30)',is_quarterly=False,add_data=True)
dv.add_formula('hml_r_std_Nm','high_r_std_Nm-low_r_std_Nm',is_quarterly=False,add_data=True)
dv.get_ts('hml_r_std_Nm').tail()
```

### 计算hpl_r_std_Nm因子

```python
dv.add_formula('hpl_r_std_Nm','high_r_std_Nm+low_r_std_Nm',is_quarterly=False,add_data=True)
dv.get_ts('hpl_r_std_Nm').tail()
```

### 计算pb_industry因子

```python
industry_list=list(dict_classify.values())
industry_code=dict()
for i in industry_list:
    industry_code[i]=sw1[sw1==i].dropna(axis=1,how='all').columns.tolist()
import pandas as pd
industry_PB=dict()
for i in industry_list:
    industry_PB[i]=pd.DataFrame(dv.get_ts('pb')[industry_code[i]])
industry_PB_mean=pd.DataFrame(columns=industry_list)
for i in industry_PB.keys():
    industry_PB_mean[i]=industry_PB[i].median(axis=1)
pb_industry=pd.DataFrame(columns=dv.get_ts('pb').columns)
for i in industry_code.keys():
    for j in industry_code[i]:
        pb_industry[j]=dv.get_ts('pb')[j]/industry_PB_mean[i]
dv.append_df(pb_industry,'pb_industry')
```

```python
import numpy as np

# 定义可买卖条件——未停牌、未涨跌停
def limit_up_down():
    trade_status = dv.get_ts('trade_status').fillna(0)
    mask_sus = trade_status == 0
    # 涨停
    up_limit = dv.add_formula('up_limit', '(close - Delay(close, 1)) / Delay(close, 1) > 0.095', is_quarterly=False)
    # 跌停
    down_limit = dv.add_formula('down_limit', '(close - Delay(close, 1)) / Delay(close, 1) < -0.095', is_quarterly=False)
    can_enter = np.logical_and(up_limit < 1, ~mask_sus) # 未涨停未停牌
    can_exit = np.logical_and(down_limit < 1, ~mask_sus) # 未跌停未停牌
    return can_enter,can_exit
```

```python
can_enter,can_exit = limit_up_down()
```

## 步骤六： 分析因子周期特点

```python
import matplotlib.pyplot as plt
from jaqs_fxdayu.research import SignalDigger
from jaqs_fxdayu.research.signaldigger import analysis

def cal_obj(signal, name, period, quantile):
#     price_bench = dv.data_benchmark
    obj = SignalDigger(output_folder="hs300/%s" % name,
                       output_format='pdf')
    obj.process_signal_before_analysis(signal,
                                   price=price,
                                   n_quantiles=quantile, period=period,
                                   mask=mask,
                                   group=sw1,
                                   can_enter = enter,
                                   can_exit = exit,
                                   commission = 0.0008
                                   )
    obj.create_full_report()
    return obj

def plot_pfm(signal, name, period=5, quantile=5):
    obj = cal_obj(signal, name, period, quantile)
    plt.show()
def signal_data(signal, name, period=5, quantile=5):
    obj = cal_obj(signal, name, period, quantile)
    return obj.signal_data
```

```python
signals_dict = {a:signal_data(neutralize_dict[a], a, 20) for a in alpha_signal} 
```

```python
ic_pn = pd.Panel({a: analysis.ic_stats(signals_dict[a]) for a in signals_dict.keys()})
```

```python
alpha_performance = round(ic_pn.minor_xs('return_ic'),2)
print(alpha_performance)
```

```python
alpha_IR = alpha_performance.loc["Ann. IR"]
alpha_IC = alpha_performance.loc["IC Mean"]
```

```python
good_alpha = alpha_IC[(abs(alpha_IC)>=0.03) & (abs(alpha_IR)>=0.25)]
```

```python
good_alpha_dict = {g: float('%.2f' % good_alpha[g]) for g in good_alpha.index}
```

## 步骤七：查看因子行业特点（最优周期）
选择最优的持有周期三年（750天）的平均行业IC，再求其平均IC，输出IC大于0.05与小于-0.05的行业

```python
signal_dict = {alpha : signal_data(dv.get_ts(alpha), alpha, period=20, quantile=5) for alpha in good_alpha.index}
```

```python
def ic_length(signal, days=750):
    return signal.loc[signal.index.levels[0][-days]:]
```

```python
from jaqs.research.signaldigger import performance as pfm

performance_dict = {}
for alpha in good_alpha.index:
    ic = pfm.calc_signal_ic(ic_length(signal_dict[alpha]), by_group=True)
    mean_ic_by_group = pfm.mean_information_coefficient(ic, by_group=True)
    performance_dict[alpha] = round(mean_ic_by_group,2)
```

```python
ic_industry = pd.Panel(performance_dict).minor_xs('ic')
```

```python
High_IC_Industry = pd.DataFrame([ic_industry[abs(ic_industry)>=0.05][alpha].dropna(how='all') for alpha in good_alpha.index]).T
```
